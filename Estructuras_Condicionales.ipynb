```python
# ==========================================================
# Calcular el total a pagar por la compra de llantas
# ==========================================================

def leer_cantidad_llantas():
    """
    Solicita la cantidad de llantas y valida que sea un número entero positivo.
    """
    while True:
        try:
            cantidad = int(input("Ingrese la cantidad de llantas: "))
            if cantidad > 0:
                return cantidad
            else:
                print("x La cantidad debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def calcular_total(cantidad):
    """
    Determina el precio por llanta y calcula el total.
    """
    if cantidad < 5:
        precio = 150
    else:
        precio = 120

    total = cantidad * precio
    return precio, total

print("=" * 50)
print("        LLANTERA - CÁLCULO DE PAGO")
print("=" * 50)

cantidad = leer_cantidad_llantas()

precio, total = calcular_total(cantidad)

print("\n========== RESULTADO ==========")
print(f"Cantidad de llantas : {cantidad}")
print(f"Precio por llanta   : S/ {precio:.2f}")
print(f"Total a pagar       : S/ {total:.2f}")
print("=" * 50)
```

```python
# ==========================================================
# Determinar el número mayor y el número menor de tres números
# ==========================================================

def leer_numero(mensaje):
    """
    Solicita un número y valida que sea correcto.
    """
    while True:
        try:
            numero = float(input(mensaje))
            return numero
        except ValueError:
            print("❌ Error: Ingrese un número válido.")


def obtener_mayor_menor(lista_numeros):
    """
    Determina el número mayor y menor de una lista.
    """
    mayor = max(lista_numeros)
    menor = min(lista_numeros)
    return mayor, menor


print("=" * 55)
print("     DETERMINAR EL NÚMERO MAYOR Y EL MENOR")
print("=" * 55)

numeros = []

for i in range(1, 4):
    numero = leer_numero(f"Ingrese el número {i}: ")
    numeros.append(numero)

mayor, menor = obtener_mayor_menor(numeros)

print("\n========== RESULTADOS ==========")
print(f"Números ingresados : {numeros}")
print(f"Número mayor       : {mayor}")
print(f"Número menor       : {menor}")
print("=" * 55)
```

```python
# ==========================================================
# Determinar si una persona debe tributar y calcular el impuesto
# ==========================================================

def leer_edad():
    """
    Solicita y valida la edad del usuario.
    """
    while True:
        try:
            edad = int(input("Ingrese su edad: "))
            if edad > 0:
                return edad
            else:
                print("x La edad debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_ingreso():
    """
    Solicita y valida el ingreso mensual del usuario.
    """
    while True:
        try:
            ingreso = float(input("Ingrese su ingreso mensual (S/): "))
            if ingreso >= 0:
                return ingreso
            else:
                print("x El ingreso no puede ser negativo.")
        except ValueError:
            print("x Error: Ingrese un monto válido.")


def calcular_impuesto(edad, ingreso):
    """
    Determina si el usuario debe tributar y calcula el impuesto.
    """
    if edad >= 18 and ingreso >= 3000:
        impuesto = ingreso * 0.05
        return True, impuesto
    else:
        return False, 0

print("=" * 60)
print("      CÁLCULO DE IMPUESTO A TRIBUTAR")
print("=" * 60)

edad = leer_edad()
ingreso = leer_ingreso()

tributa, impuesto = calcular_impuesto(edad, ingreso)

print("\n========== RESULTADOS ==========")
print(f"Edad               : {edad} años")
print(f"Ingreso mensual    : S/ {ingreso:.2f}")

if tributa:
    print("Estado             : DEBE TRIBUTAR")
    print(f"Impuesto (5%)      : S/ {impuesto:.2f}")
else:
    print("Estado             : NO DEBE TRIBUTAR")

print("=" * 60)
```

```python
# =========================================================
# Cálculo de la colegiatura de un alumno
# ==========================================================

COSTO_MATERIA = 100


def leer_materias():
    """
    Solicita y valida la cantidad de materias.
    """
    while True:
        try:
            materias = int(input("Ingrese la cantidad de materias: "))
            if materias > 0:
                return materias
            else:
                print("x La cantidad de materias debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_promedio():
    """
    Solicita y valida el promedio del alumno.
    """
    while True:
        try:
            promedio = float(input("Ingrese el promedio del último periodo (0 - 20): "))
            if 0 <= promedio <= 20:
                return promedio
            else:
                print("x El promedio debe estar entre 0 y 20.")
        except ValueError:
            print("x Error: Ingrese un número válido.")


def calcular_colegiatura(materias, promedio):
    """
    Calcula el monto a pagar considerando el descuento.
    """
    costo_total = materias * COSTO_MATERIA

    if promedio >= 18:
        materias_descuento = materias // 2
        descuento = materias_descuento * COSTO_MATERIA * 0.50
    else:
        descuento = 0

    total_pagar = costo_total - descuento

    return costo_total, descuento, total_pagar


print("=" * 60)
print("      CÁLCULO DE LA COLEGIATURA")
print("=" * 60)

materias = leer_materias()
promedio = leer_promedio()

costo, descuento, total = calcular_colegiatura(materias, promedio)


print(f"Cantidad de materias : {materias}")
print(f"Costo por materia    : S/ {COSTO_MATERIA:.2f}")
print(f"Costo total          : S/ {costo:.2f}")
print(f"Promedio             : {promedio:.2f}")
print(f"Descuento            : S/ {descuento:.2f}")
print(f"Total a pagar        : S/ {total:.2f}")
print("=" * 60)
```

```python
# ==========================================================
# Determinar el tipo de triángulo
# Equilátero: Mostrar el perímetro
# Isósceles: Mostrar los lados iguales
# Escaleno: Mostrar el área
# ==========================================================

import math


def leer_lado(nombre):
    """
    Solicita y valida un lado del triángulo.
    """
    while True:
        try:
            lado = float(input(f"Ingrese el lado {nombre}: "))
            if lado > 0:
                return lado
            else:
                print("x El lado debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número válido.")


def es_triangulo(a, b, c):
    """
    Verifica si los lados forman un triángulo válido.
    """
    return (a + b > c) and (a + c > b) and (b + c > a)


def tipo_triangulo(a, b, c):
    """
    Determina el tipo de triángulo.
    """
    if a == b == c:
        return "Equilátero"
    elif a == b or a == c or b == c:
        return "Isósceles"
    else:
        return "Escaleno"


def calcular_area(a, b, c):
    """
    Calcula el área utilizando la fórmula de Herón.
    """
    s = (a + b + c) / 2
    return math.sqrt(s * (s - a) * (s - b) * (s - c))

print("=" * 60)
print("       CLASIFICACIÓN DE TRIÁNGULOS")
print("=" * 60)

lado_a = leer_lado("A")
lado_b = leer_lado("B")
lado_c = leer_lado("C")

if es_triangulo(lado_a, lado_b, lado_c):

    tipo = tipo_triangulo(lado_a, lado_b, lado_c)

    print("\n========== RESULTADOS ==========")
    print(f"Lado A : {lado_a}")
    print(f"Lado B : {lado_b}")
    print(f"Lado C : {lado_c}")
    print(f"Tipo   : {tipo}")

    if tipo == "Equilátero":
        perimetro = lado_a + lado_b + lado_c
        print(f"Perímetro: {perimetro:.2f}")

    elif tipo == "Isósceles":

        if lado_a == lado_b:
            print("Lados iguales: A y B")
        elif lado_a == lado_c:
            print("Lados iguales: A y C")
        else:
            print("Lados iguales: B y C")

    else:
        area = calcular_area(lado_a, lado_b, lado_c)
        print(f"Área del triángulo: {area:.2f}")

else:
    print("\n Los valores ingresados NO forman un triángulo válido.")

print("=" * 60)
```

```python
# =========================================================
# Mostrar la hora N segundos después
# ==========================================================

def leer_hora():
    """
    Solicita y valida la hora (0 - 23).
    """
    while True:
        try:
            hora = int(input("Ingrese la hora (0-23): "))
            if 0 <= hora <= 23:
                return hora
            else:
                print("x La hora debe estar entre 0 y 23.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_minuto():
    """
    Solicita y valida los minutos (0 - 59).
    """
    while True:
        try:
            minuto = int(input("Ingrese los minutos (0-59): "))
            if 0 <= minuto <= 59:
                return minuto
            else:
                print("x Los minutos deben estar entre 0 y 59.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_segundo():
    """
    Solicita y valida los segundos (0 - 59).
    """
    while True:
        try:
            segundo = int(input("Ingrese los segundos (0-59): "))
            if 0 <= segundo <= 59:
                return segundo
            else:
                print("x Los segundos deben estar entre 0 y 59.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_n_segundos():
    """
    Solicita y valida la cantidad de segundos a agregar.
    """
    while True:
        try:
            n = int(input("Ingrese los segundos a sumar: "))
            if n >= 0:
                return n
            else:
                print("x Debe ingresar un valor mayor o igual a cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def calcular_nueva_hora(hora, minuto, segundo, n):
    """
    Calcula la nueva hora después de sumar N segundos.
    """
    total_segundos = hora * 3600 + minuto * 60 + segundo
    total_segundos += n

    # Mantener la hora dentro de un día (24 horas)
    total_segundos %= 24 * 3600

    nueva_hora = total_segundos // 3600
    resto = total_segundos % 3600
    nuevo_minuto = resto // 60
    nuevo_segundo = resto % 60

    return nueva_hora, nuevo_minuto, nuevo_segundo


print("=" * 60)
print("      CÁLCULO DE LA HORA N SEGUNDOS DESPUÉS")
print("=" * 60)

hora = leer_hora()
minuto = leer_minuto()
segundo = leer_segundo()
n = leer_n_segundos()

h, m, s = calcular_nueva_hora(hora, minuto, segundo, n)

print(f"Hora ingresada : {hora:02d}:{minuto:02d}:{segundo:02d}")
print(f"Segundos sumados: {n}")
print(f"Nueva hora     : {h:02d}:{m:02d}:{s:02d}")
print("=" * 60)
```

```python
# ==========================================================
# Calcular el pago de una llamada telefónica
# ==========================================================

def leer_minutos():
    """
    Solicita y valida la duración de la llamada en minutos.
    """
    while True:
        try:
            minutos = int(input("Ingrese la duración de la llamada (minutos): "))
            if minutos > 0:
                return minutos
            else:
                print("x Los minutos deben ser mayores que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_dia():
    """
    Solicita el día de la llamada.
    """
    while True:
        dia = input("¿La llamada fue en domingo? (S/N): ").strip().upper()
        if dia in ("S", "N"):
            return dia
        print("x Responda únicamente con S o N.")


def leer_turno():
    """
    Solicita el turno únicamente si la llamada fue en domingo.
    """
    while True:
        turno = input("¿La llamada fue en turno nocturno? (S/N): ").strip().upper()
        if turno in ("S", "N"):
            return turno
        print("x Responda únicamente con S o N.")


def calcular_costo(minutos):
    """
    Calcula el costo de la llamada según la tarifa.
    """
    costo = 0

    for minuto in range(1, minutos + 1):

        if minuto <= 5:
            costo += 1.00

        elif minuto <= 8:
            costo += 0.80

        elif minuto <= 10:
            costo += 0.70

        else:
            costo += 0.50

    return costo


# ===========================

print("=" * 65)
print("         CÁLCULO DEL COSTO DE UNA LLAMADA")
print("=" * 65)

minutos = leer_minutos()

subtotal = calcular_costo(minutos)

dia = leer_dia()

impuesto = 0

if dia == "S":

    turno = leer_turno()

    if turno == "S":
        impuesto = subtotal * 0.05
    else:
        impuesto = subtotal * 0.03

total = subtotal + impuesto

print("\n============== RESULTADOS ==============")
print(f"Duración de la llamada : {minutos} minutos")
print(f"Subtotal               : S/ {subtotal:.2f}")
print(f"Impuesto               : S/ {impuesto:.2f}")
print(f"TOTAL A PAGAR          : S/ {total:.2f}")
print("=" * 65)
```

```python
# ==========================================================
# EJERCICIO 8
# Ordenar tres números de forma ascendente o descendente
# ==========================================================

def leer_numero(mensaje):
    """
    Solicita y valida un número entero.
    """
    while True:
        try:
            numero = int(input(mensaje))
            return numero
        except ValueError:
            print(" Error: Ingrese un número entero válido.")


def leer_orden():
    """
    Solicita el tipo de orden.
    """
    while True:
        opcion = input("Orden (A = Ascendente / D = Descendente): ").strip().upper()

        if opcion in ("A", "D"):
            return opcion

        print("x Opción no válida. Ingrese A o D.")

print("=" * 60)
print("      ORDENAR TRES NÚMEROS")
print("=" * 60)

while True:

    numeros = []

    for i in range(1, 4):
        numero = leer_numero(f"Ingrese el número {i}: ")
        numeros.append(numero)

    # Validar que los tres números sean diferentes
    if len(set(numeros)) == 3:
        break
    else:
        print("\nx Los números deben ser diferentes.")
        print("Ingrese nuevamente los datos.\n")

orden = leer_orden()

if orden == "A":
    numeros.sort()
    tipo = "ASCENDENTE"
else:
    numeros.sort(reverse=True)
    tipo = "DESCENDENTE"

print("\n========== RESULTADOS ==========")
print(f"Orden seleccionado : {tipo}")
print("Números ordenados  :", numeros)
print("=" * 60)
```

```python
# ==========================================================
# EJERCICIO 9
# Calcular el total a pagar por la compra de camisas
# ==========================================================

def leer_cantidad_camisas():
    """
    Solicita y valida la cantidad de camisas.
    """
    while True:
        try:
            cantidad = int(input("Ingrese la cantidad de camisas: "))
            if cantidad > 0:
                return cantidad
            else:
                print("x La cantidad debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_precio_camisa():
    """
    Solicita y valida el precio de cada camisa.
    """
    while True:
        try:
            precio = float(input("Ingrese el precio de cada camisa (S/): "))
            if precio > 0:
                return precio
            else:
                print("x El precio debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un precio válido.")


def calcular_compra(cantidad, precio):
    """
    Calcula el subtotal, descuento y total a pagar.
    """
    subtotal = cantidad * precio

    if 1 <= cantidad <= 4:
        porcentaje_descuento = 12.5
    elif 5 <= cantidad <= 8:
        porcentaje_descuento = 20
    else:
        porcentaje_descuento = 31.5

    descuento = subtotal * (porcentaje_descuento / 100)
    total = subtotal - descuento

    return subtotal, porcentaje_descuento, descuento, total


print("=" * 60)
print("      CÁLCULO DEL TOTAL POR COMPRA DE CAMISAS")
print("=" * 60)

cantidad = leer_cantidad_camisas()
precio = leer_precio_camisa()

subtotal, porcentaje, descuento, total = calcular_compra(cantidad, precio)

print("\n========== RESULTADOS ==========")
print(f"Cantidad de camisas      : {cantidad}")
print(f"Precio por camisa        : S/ {precio:.2f}")
print(f"Compra sin descuento     : S/ {subtotal:.2f}")
print(f"Porcentaje de descuento  : {porcentaje:.1f}%")
print(f"Monto del descuento      : S/ {descuento:.2f}")
print(f"Compra con descuento     : S/ {total:.2f}")
print("=" * 60)
```

```python
# ==========================================================
# Cálculo del costo de una llamada internacional
# ==========================================================

ZONAS = {
    10: ("América del Norte", 2.20),
    12: ("América Central", 2.50),
    20: ("América del Sur", 1.20),
    22: ("Asia", 3.50),
    30: ("Europa", 3.00),
    32: ("África", 3.20)
}


# ==========================================================
# EJERCICIO 10
# Cálculo del costo de una llamada internacional
# ==========================================================

# Diccionario con las zonas geográficas y el costo por minuto
ZONAS = {
    10: ("América del Norte", 2.20),
    12: ("América Central", 2.50),
    20: ("América del Sur", 1.20),
    22: ("Asia", 3.50),
    30: ("Europa", 3.00),
    32: ("África", 3.20)
}


def leer_codigo():
    """
    Solicita y valida el código de la zona.
    """
    while True:
        try:
            codigo = int(input("Ingrese el código de la zona (10,12,20,22,30,32): "))

            if codigo in ZONAS:
                return codigo

            print("x Código de zona no válido.")

        except ValueError:
            print("x Error: Ingrese un número entero.")


def leer_minutos():
    """
    Solicita y valida la cantidad de minutos.
    """
    while True:
        try:
            minutos = int(input("Ingrese la duración de la llamada (minutos): "))

            if minutos > 0:
                return minutos

            print("x Los minutos deben ser mayores que cero.")

        except ValueError:
            print("x Error: Ingrese un número entero.")


def calcular_costo(codigo, minutos):
    """
    Calcula el costo total de la llamada.
    """
    zona, tarifa = ZONAS[codigo]
    total = minutos * tarifa

    return zona, tarifa, total


print("=" * 65)
print("     CÁLCULO DEL COSTO DE LLAMADAS INTERNACIONALES")
print("=" * 65)

codigo = leer_codigo()
minutos = leer_minutos()

zona, tarifa, total = calcular_costo(codigo, minutos)

print("\n============== RESULTADOS ==============")
print(f"Código de zona      : {codigo}")
print(f"Zona geográfica     : {zona}")
print(f"Costo por minuto    : S/ {tarifa:.2f}")
print(f"Duración llamada    : {minutos} minutos")
print(f"Costo total         : S/ {total:.2f}")
print("=" * 65)
```
