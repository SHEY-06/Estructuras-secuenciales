```python
# Determina la velocidad final de un auto si su velocidad inicial es 0 km/h y su aceleración es 0.8 m/s2; y el tiempo transcurrido es de 30 segundos.

def leer_float(msg,minimo=None):
    while True:
        try:
            x=float(input(msg))
            if minimo is not None and x<minimo:
                print("Ingrese un valor válido.")
                continue
            return x
        except:
            print("Dato inválido.")
vi=leer_float("Velocidad inicial (m/s): ",0)
a=leer_float("Aceleración (m/s²): ",0)
t=leer_float("Tiempo (s): ",0)
vf=vi+a*t
print(f"Velocidad final: {vf:.2f} m/s")
```

```python
#Un estudiante realiza cuatro exámenes, los cuales tienen la misma ponderación. Representen el algoritmo correspondiente para obtener el promedio de las calificaciones obtenidas
notas=[]
for i in range(4):
    while True:
        try:
            n=float(input(f"Nota {i+1}: "))
            if 0<=n<=20:
                notas.append(n);break
            print("La nota debe estar entre 0 y 20.")
        except:
            print("Dato inválido.")
prom=sum(notas)/len(notas)
print(f"Promedio: {prom:.2f}")
print("Aprobado" if prom>=11 else "Desaprobado")
```

```python
#Tres personas deciden invertir su dinero para fundar una empresa. Cada una de ellas invierte una cantidad distinta. Obtener el porcentaje que cada uno invierte con respecto a la cantidad total invertida.
inv=[]
for i in range(3):
    while True:
        try:
            x=float(input(f"Inversión persona {i+1}: "))
            if x>0:
                inv.append(x);break
            print("Debe ser mayor que cero.")
        except:
            print("Dato inválido.")
total=sum(inv)
for i,v in enumerate(inv,1):
    print(f"Persona {i}: {v/total*100:.2f}%")
```

```python
#Un vendedor recibe su sueldo base más 25% extra por comisión de sus ventas; el vendedor desea saber cuánto dinero recibirá por concepto de
#comisiones por lastres ventas que realiza en el mes y el total que recibirá en el mes tomando en cuenta sueldo base y comisiones.

sueldo=float(input("Sueldo base: "))
ventas=[]
for i in range(3):
    ventas.append(float(input(f"Venta {i+1}: ")))
comision=sum(ventas)*0.25
print(f"Comisión: S/ {comision:.2f}")
print(f"Total: S/ {sueldo+comision:.2f}")
```

```python
#Suponga que un individuo desea invertir su capital en un banco y desea saber cuánto dinero ganará después de un año si el banco paga a razón de 2% mensual.
capital=float(input("Capital inicial: "))
monto=capital
for _ in range(12):
    monto*=1.02
print(f"Monto final: S/ {monto:.2f}")
print(f"Ganancia: S/ {monto-capital:.2f}")
```

```python
#Un estacionamiento requiere determinar el cobro que debe aplicar a las
#personas que lo utilizan. Considere que el cobro es con base en las
#horas que lo disponen y que las fracciones de hora se toman como
#completas y realice un programa que permita determinar el cobro. El
#pago por hora es de S/5.00 y el pago por los minutos restantes de S/0.10
#por minuto.

# Considerando fracción de hora como completa

horas=int(input("Horas completas: "))
minutos=int(input("Minutos adicionales: "))
if minutos<0 or minutos>=60:
    print("Minutos inválidos.")
else:
    total_horas = horas
    if minutos > 0:
        total_horas += 1
    total=total_horas*5
    print(f"Pago: S/ {total:.2f}")
```

```python
#Una empresa constructora vende terrenos con la forma A de la figura.
#Obtener el área respectiva de un terreno de medidas ingresando el valor
#de los 3 lados.

def leer_dato(nombre):
    """Solicita un número positivo."""
    while True:
        try:
            valor = float(input(f"Ingrese {nombre}: "))
            if valor > 0:
                return valor
            else:
                print("x El valor debe ser mayor que cero.")
        except ValueError:
            print("x Ingrese un número válido.")


def calcular_area(a, b, c):
    """Calcula el área de la figura A."""
    area_rectangulo = b * c
    area_triangulo = (b * (a - c)) / 2
    return area_rectangulo + area_triangulo


A = leer_dato("lado A")
B = leer_dato("lado B")
C = leer_dato("lado C")

# Validación
if A <= C:
    print("\nx Error: El lado A debe ser mayor que C.")
else:
    area = calcular_area(A, B, C)

    print(f"Lado A : {A:.2f}")
    print(f"Lado B : {B:.2f}")
    print(f"Lado C : {C:.2f}")
    print(f"Área del terreno: {area:.2f} m²")
```

```python
#Cuanto pagaría un cliente por una compra con descuento y cuanto sería
#el cambio que recibiría en un pago en efectivo. Ingresar el monto base,
#el porcentaje de descuento y el con cuánto dinero está pagando el
#cliente.
monto=float(input("Monto: "))
desc=float(input("% descuento: "))
pago=float(input("Pago: "))
total=monto-(monto*desc/100)
if pago>=total:
    print(f"Total: S/ {total:.2f}")
    print(f"Cambio: S/ {pago-total:.2f}")
else:
    print(f"Faltan S/ {total-pago:.2f}")
```

```python
#Se necesita saber la cantidad de dinero que un cajero automático debe
#proporcionar en diferentes denominaciones, teniendo en cuenta que el
#usuario debe ingresar la cantidad a retirar y el cajero tiene disponible las
#siguientes denominaciones de dinero: 200, 100, 50, 20, 10, 5, 1. Se debe
#buscar entregar la menor cantidad de dinero por parte del cajero.

monto=int(input("Monto a retirar: "))
den=[200,100,50,20,10,5,1]
for d in den:
    c=monto//d
    monto%=d
    print(f"{d}: {c}")
```
