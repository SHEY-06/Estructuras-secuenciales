```python
# ==========================================================
# Calcular la suma y el promedio de los números
# pares e impares desde 1 hasta N
# ==========================================================

def leer_numero():
    """
    Solicita y valida un número entero positivo.
    """
    while True:
        try:
            n = int(input("Ingrese un número entero positivo (N): "))
            if n > 0:
                return n
            else:
                print("x El número debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def calcular_pares_impares(n):
    """
    Calcula la suma y el promedio de números pares e impares.
    """
    suma_pares = 0
    suma_impares = 0
    cantidad_pares = 0
    cantidad_impares = 0

    for numero in range(1, n + 1):

        if numero % 2 == 0:
            suma_pares += numero
            cantidad_pares += 1
        else:
            suma_impares += numero
            cantidad_impares += 1

    promedio_pares = (
        suma_pares / cantidad_pares if cantidad_pares > 0 else 0
    )

    promedio_impares = (
        suma_impares / cantidad_impares if cantidad_impares > 0 else 0
    )

    return (
        suma_pares,
        promedio_pares,
        suma_impares,
        promedio_impares
    )


print("=" * 60)
print("     SUMA Y PROMEDIO DE NÚMEROS PARES E IMPARES")
print("=" * 60)

n = leer_numero()

suma_pares, promedio_pares, suma_impares, promedio_impares = calcular_pares_impares(n)

print("\n========== RESULTADOS ==========")
print(f"Rango evaluado           : 1 hasta {n}")
print(f"Suma de números pares    : {suma_pares}")
print(f"Promedio de pares        : {promedio_pares:.2f}")
print(f"Suma de números impares  : {suma_impares}")
print(f"Promedio de impares      : {promedio_impares:.2f}")
print("=" * 60)
```

```python
# ==========================================================
# Determinar cuántos números son:
# Pares, Impares, Positivos, Negativos y Neutros
# ==========================================================

def leer_numero(posicion):
    """
    Solicita y valida un número entero.
    """
    while True:
        try:
            numero = int(input(f"Ingrese el número {posicion}: "))
            return numero
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def clasificar_numeros(lista):
    """
    Clasifica los números según sus características.
    """
    pares = 0
    impares = 0
    positivos = 0
    negativos = 0
    neutros = 0

    for numero in lista:

        # Pares e impares
        if numero % 2 == 0:
            pares += 1
        else:
            impares += 1

        # Positivos, negativos y neutros
        if numero > 0:
            positivos += 1
        elif numero < 0:
            negativos += 1
        else:
            neutros += 1

    return pares, impares, positivos, negativos, neutros


print("=" * 65)
print("      CLASIFICACIÓN DE 10 NÚMEROS")
print("=" * 65)

numeros = []

for i in range(1, 11):
    numeros.append(leer_numero(i))

pares, impares, positivos, negativos, neutros = clasificar_numeros(numeros)

print("\n============== RESULTADOS ==============")
print(f"Números ingresados : {numeros}")
print(f"Cantidad de pares      : {pares}")
print(f"Cantidad de impares    : {impares}")
print(f"Cantidad de positivos  : {positivos}")
print(f"Cantidad de negativos  : {negativos}")
print(f"Cantidad de neutros    : {neutros}")
print("=" * 65)
```

```python
# ==========================================================
# Determinar simultáneamente el número máximo y mínimo
# de una lista de N números
# ==========================================================

def leer_cantidad():
    """
    Solicita y valida la cantidad de números.
    """
    while True:
        try:
            n = int(input("Ingrese la cantidad de números (N): "))
            if n > 0:
                return n
            else:
                print("x La cantidad debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_numero(posicion):
    """
    Solicita y valida un número.
    """
    while True:
        try:
            return float(input(f"Ingrese el número {posicion}: "))
        except ValueError:
            print("x Error: Ingrese un número válido.")


def obtener_maximo_minimo(numeros):
    """
    Determina simultáneamente el número mayor y menor.
    """
    mayor = numeros[0]
    menor = numeros[0]

    for numero in numeros[1:]:

        if numero > mayor:
            mayor = numero

        if numero < menor:
            menor = numero

    return mayor, menor


print("=" * 60)
print("     DETERMINAR EL MÁXIMO Y EL MÍNIMO")
print("=" * 60)

n = leer_cantidad()

lista_numeros = []

for i in range(1, n + 1):
    lista_numeros.append(leer_numero(i))

mayor, menor = obtener_maximo_minimo(lista_numeros)

print("\n========== RESULTADOS ==========")
print(f"Números ingresados : {lista_numeros}")
print(f"Número máximo      : {mayor}")
print(f"Número mínimo      : {menor}")
print("=" * 60)
```

```python
# ==========================================================
# Calcular el ahorro acumulado durante un año
# ==========================================================

MESES = [
    "Enero", "Febrero", "Marzo", "Abril",
    "Mayo", "Junio", "Julio", "Agosto",
    "Septiembre", "Octubre", "Noviembre", "Diciembre"
]


def leer_deposito(mes):
    """
    Solicita y valida el depósito correspondiente a un mes.
    """
    while True:
        try:
            deposito = float(input(f"Ingrese el depósito de {mes} (S/): "))
            if deposito >= 0:
                return deposito
            else:
                print("x El depósito no puede ser negativo.")
        except ValueError:
            print("x Error: Ingrese un monto válido.")


def calcular_ahorro(depositos):
    """
    Calcula el ahorro acumulado mes a mes.
    """
    acumulado = 0

    print("\n=========== AHORRO ACUMULADO ===========")

    for i in range(len(depositos)):
        acumulado += depositos[i]
        print(f"{MESES[i]:<12} -> Depósito: S/ {depositos[i]:7.2f} | Acumulado: S/ {acumulado:8.2f}")

    return acumulado
1

print("=" * 65)
print("      AHORRO ACUMULADO DURANTE 12 MESES")
print("=" * 65)

depositos = []

for mes in MESES:
    deposito = leer_deposito(mes)
    depositos.append(deposito)

ahorro_total = calcular_ahorro(depositos)

print("\n============== RESULTADO FINAL ==============")
print(f"Ahorro total en el año: S/ {ahorro_total:.2f}")
print("=" * 65)
```

```python
# ==========================================================
# Simulación de un cronómetro (HH:MM:SS)
# ==========================================================

import time


def leer_tiempo():
    """
    Solicita y valida el tiempo de duración del cronómetro.
    """
    while True:
        try:
            horas = int(input("Ingrese las horas: "))
            minutos = int(input("Ingrese los minutos: "))
            segundos = int(input("Ingrese los segundos: "))

            if horas >= 0 and 0 <= minutos < 60 and 0 <= segundos < 60:
                return horas, minutos, segundos
            else:
                print("x Minutos y segundos deben estar entre 0 y 59.")
        except ValueError:
            print("x Error: Ingrese números enteros válidos.")


def cronometro(horas, minutos, segundos):
    """
    Simula el funcionamiento de un cronómetro.
    """
    total_segundos = horas * 3600 + minutos * 60 + segundos

    print("\n========== CRONÓMETRO ==========\n")

    for tiempo in range(total_segundos + 1):

        h = tiempo // 3600
        m = (tiempo % 3600) // 60
        s = tiempo % 60

        print(f"\r⏱ {h:02d}:{m:02d}:{s:02d}", end="")

        # Simulación referencial (1 segundo)
        time.sleep(1)

    print("\n\nCronómetro finalizado.")

print("=" * 60)
print("        CRONÓMETRO DIGITAL")
print("=" * 60)

horas, minutos, segundos = leer_tiempo()

cronometro(horas, minutos, segundos)

print("=" * 60)
```

```python
# ==========================================================
# Calcular el factorial de dos números y mostrar
# hasta qué valor ambos factoriales son iguales.
# ==========================================================

def leer_numero(mensaje):
    """
    Solicita y valida un número entero positivo.
    """
    while True:
        try:
            numero = int(input(mensaje))
            if numero >= 0:
                return numero
            else:
                print("x El número debe ser mayor o igual a cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def calcular_factorial(n):
    """
    Calcula el factorial de un número.
    """
    factorial = 1

    if n == 0 or n == 1:
        return 1

    for i in range(2, n + 1):
        factorial *= i

    return factorial


def desarrollo_factorial(n):
    """
    Devuelve el desarrollo del factorial.
    """
    if n == 0:
        return "1"

    return " x ".join(str(i) for i in range(1, n + 1))

print("=" * 65)
print("        FACTORIAL DE DOS NÚMEROS")
print("=" * 65)

num1 = leer_numero("Ingrese el primer número: ")
num2 = leer_numero("Ingrese el segundo número: ")

factorial1 = calcular_factorial(num1)
factorial2 = calcular_factorial(num2)

print("\n========== DESARROLLO ==========")
print(f"{num1}! = {desarrollo_factorial(num1)}")
print(f"{num2}! = {desarrollo_factorial(num2)}")

print("\n========== RESULTADOS ==========")
print(f"{num1}! = {factorial1}")
print(f"{num2}! = {factorial2}")

# Determinar hasta qué valor son iguales
limite = min(num1, num2)

print(f"\nAmbos factoriales son iguales hasta el múltiplo {limite}.")
print(f"Desarrollo común: {desarrollo_factorial(limite)}")

print("=" * 65)
```

```python
# ==========================================================
# Registro de asistentes a una fiesta
# ==========================================================

def leer_edad():
    """
    Solicita y valida la edad del asistente.
    """
    while True:
        try:
            edad = int(input("Ingrese la edad: "))
            if edad > 0:
                return edad
            else:
                print("x La edad debe ser mayor que cero.")
        except ValueError:
            print("x Error: Ingrese un número entero válido.")


def leer_sexo():
    """
    Solicita y valida el sexo del asistente.
    """
    while True:
        sexo = input("Ingrese el sexo (M = Masculino / F = Femenino): ").strip().upper()

        if sexo in ("M", "F"):
            return sexo

        print("x Opción no válida. Ingrese M o F.")


def registrar_otro():
    """
    Pregunta si desea registrar otro asistente.
    """
    while True:
        opcion = input("¿Desea registrar otra persona? (S/N): ").strip().upper()

        if opcion in ("S", "N"):
            return opcion

        print("x Responda S o N.")


def calcular_estadisticas(asistentes):
    """
    Calcula las estadísticas de los asistentes.
    """
    total = len(asistentes)

    hombres = 0
    mujeres = 0

    suma_hombres = 0
    suma_mujeres = 0

    edad_mayor = asistentes[0]["edad"]
    edad_menor = asistentes[0]["edad"]

    for persona in asistentes:

        edad = persona["edad"]
        sexo = persona["sexo"]

        # Contar hombres y mujeres
        if sexo == "M":
            hombres += 1
            suma_hombres += edad
        else:
            mujeres += 1
            suma_mujeres += edad

        # Edad mayor
        if edad > edad_mayor:
            edad_mayor = edad

        # Edad menor
        if edad < edad_menor:
            edad_menor = edad

    promedio_hombres = (
        suma_hombres / hombres if hombres > 0 else 0
    )

    promedio_mujeres = (
        suma_mujeres / mujeres if mujeres > 0 else 0
    )

    return (
        total,
        hombres,
        mujeres,
        promedio_hombres,
        promedio_mujeres,
        edad_menor,
        edad_mayor
    )

print("=" * 65)
print("        REGISTRO DE ASISTENTES A UNA FIESTA")
print("=" * 65)

asistentes = []

while True:

    edad = leer_edad()
    sexo = leer_sexo()

    asistentes.append({
        "edad": edad,
        "sexo": sexo
    })

    if registrar_otro() == "N":
        break

(
    total,
    hombres,
    mujeres,
    promedio_hombres,
    promedio_mujeres,
    edad_menor,
    edad_mayor
) = calcular_estadisticas(asistentes)

print("\n============== RESULTADOS ==============")
print(f"Total de asistentes       : {total}")
print(f"Hombres                   : {hombres}")
print(f"Mujeres                   : {mujeres}")
print(f"Promedio edad hombres     : {promedio_hombres:.2f} años")
print(f"Promedio edad mujeres     : {promedio_mujeres:.2f} años")
print(f"Edad de la persona menor  : {edad_menor} años")
print(f"Edad de la persona mayor  : {edad_mayor} años")
print("=" * 65)
```

```python
# ==========================================================
# Calcular el producto:
# 100 × 98 × 96 × 94 × ... × 2
# utilizando arrays (listas)
# ==========================================================

def generar_arreglo():
    """
    Genera un arreglo con los números pares
    desde 100 hasta 2.
    """
    numeros = []

    for numero in range(100, 0, -2):
        numeros.append(numero)

    return numeros


def calcular_producto(arreglo):
    """
    Calcula el producto de todos los elementos
    del arreglo.
    """
    producto = 1

    for numero in arreglo:
        producto *= numero

    return producto

print("=" * 65)
print("      PRODUCTO DE LA SERIE UTILIZANDO ARRAYS")
print("=" * 65)

numeros = generar_arreglo()

producto = calcular_producto(numeros)

print("\n=========== SERIE ===========")
print(numeros)

print("\n========= RESULTADOS =========")
print(f"Cantidad de elementos : {len(numeros)}")
print(f"Producto total        : {producto}")
print("=" * 65)
```
